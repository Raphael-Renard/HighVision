```python
import cv2 as cv
import os
from data_retrieval.forbin import getDataset
from matplotlib import pyplot as plt
from absolute_path import absolutePath
from tqdm import tqdm

corpus = "forbin"
```

cv2 can read images that appear full white with PIL.image (with a "corrupted" warning, removed by saving the result)

```python
paths,_,_ = getDataset(mode="image")

errors = []
for i in tqdm(range(len(paths))):
    try :
        cv.imwrite(paths[i], cv.imread(paths[i]))
    except:
        errors.append(paths[i])
```

```python
_,m,_ = getDataset(mode="image")

for i in tqdm(range(len(m))):
    path = m[i]['verso']
    if path is not None:
        try :
            cv.imwrite(path, cv.imread(path))
        except:
            errors.append(path)
```

```python
for error in errors:
    path = error.split("/")
    prefix = absolutePath + "corpus/" + corpus + "/"
    os.rename(prefix + path[-2] + "/" + path[-1], prefix + "corrupted/" + path[-1])
```

```python
paths,_,_ = getDataset(mode="image")

for i in range(len(paths)):
    a = plt.imread(paths[i])
    if a.max() <= 1:
        a = (a * 255).astype('uint8')
        plt.imsave(paths[i], a)
    elif a.max() > 255:
        a = (a / a.max() * 255).astype('uint8')
        plt.imsave(paths[i], a)
```
