```python
import sys
path = "../.."
if path not in sys.path:
    sys.path.insert(0, path)
```

```python
from data_retrieval.tools.uniformize_image import uniformize_image
from data_retrieval.lipade_groundtruth import getDataset
from absolute_path import absolutePath
from PIL import Image

N = 256
```

```python
import os

x,_,_ = getDataset(mode="all")

for path in x:
    if not(os.path.exists(path)): # pictures deleted from corruption handling
        continue
    # remove "corpus/"
    relative = path[len(absolutePath):].split("/")
    relative = "/".join(relative[1:])

    im = uniformize_image(path, N)

    Image.fromarray(im).save(absolutePath + "mini_corpus/" + relative)
```
