```python
import sys
path = "../.."
if path not in sys.path:
    sys.path.insert(0, path)
```

```python
from data_retrieval.tools.uniformize_image import uniformize_image
from data_retrieval.lipade_groundtruth import getDataset
from matplotlib import pyplot as plt
from absolute_path import absolutePath
from PIL import Image
from tqdm import tqdm
import os

N = 256
```

```python
x,_,_ = getDataset(mode="similar")

for path in tqdm(x):
    if not(os.path.exists(path)): # pictures deleted from corruption handling
        continue
    # remove "corpus/"
    relativePath = path[len(absolutePath):].split("/")
    file = relativePath[-1]
    folder = "/".join(relativePath[1:-1])

    if not os.path.exists(absolutePath + "mini_corpus/" + folder):
        os.makedirs(absolutePath + "mini_corpus/" + folder)

    try:
        im = uniformize_image(path, N)
    except:
        continue

    Image.fromarray(im).save(absolutePath + "mini_corpus/" + folder + "/" + file)
```
