```python
import sys
path = ".."
if path not in sys.path:
    sys.path.insert(0, path)
```

```python
from data_retrieval import lipade_groundtruth
from absolute_path import absolutePath
import matplotlib.pyplot as plt

corpus = 'lipade_groundtruth'
```

```python
x,m,y = lipade_groundtruth.getDataset(mode="similar")
data = []
for i in range(len(x)):
    generatedPath = absolutePath + 'data_generation/generated/' + corpus + '/' + x[i].split('/')[-1].split('.')[0] + '.csv'
    data.append((x[i], m[0][i], generatedPath, y[i]))
```

```python
maxDisplayed = 99999

for g in range(max(y) + 1)[:maxDisplayed]:
    print("Group", g, ":")
    for img,meta,genPath,g_ in data:
        if g == g_:
            with open(genPath, 'r') as f:
                gen = [desc.rstrip() for desc in f.readlines()]
            if len(meta) > 0:
                print(meta)
            for prompt, generated in [line.split(';') for line in gen]:
                print(prompt, "->", generated)
            plt.imshow(plt.imread(img), cmap='grey')
            plt.show()
    break
```
